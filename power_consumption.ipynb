```python
# Create labels
y_classified = pd.cut(
    y,
    bins=[0, 1, 3, float('inf')],
    labels=['low', 'medium', 'high']
)

# Check the distribution
print(y_classified.value_counts())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Define split again
split = int(len(X) * 0.8)

# Split classified y
y_train_clf = y_classified.iloc[:split]
y_test_clf = y_classified.iloc[split:]

# Build and train
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train_clf)

# Predict
y_pred_clf = rf_model.predict(X_test)

# Evaluate
print(classification_report(y_test_clf, y_pred_clf))
```

```python
# Define tariff rates (per kWh)
def get_tariff(hour):
    if 23 <= hour or hour < 7:
        return 0.08   # cheap night rate
    elif 7 <= hour < 17:
        return 0.15   # normal day rate
    else:
        return 0.25   # expensive peak rate

# Apply tariff to test data
hours_test = X_test['hours']
tariffs = hours_test.apply(get_tariff)

# Calculate actual cost
actual_cost = (y_test * tariffs / 60).sum()

# Calculate predicted cost
predicted_series = pd.Series(y_pred_xgb, index=X_test.index)
predicted_cost = (predicted_series * tariffs / 60).sum()

print(f"Actual cost:    {actual_cost:.2f}")
print(f"Predicted cost: {predicted_cost:.2f}")
```

```python
# Optimized cost - shift peak usage to cheap rate
def get_optimized_tariff(hour, consumption):
    if consumption > 3:           # high usage
        return 0.08               # force to cheap night rate
    return get_tariff(hour)       # otherwise keep normal tariff

optimized_tariffs = pd.Series([
    get_optimized_tariff(h, c) 
    for h, c in zip(hours_test, y_test)
], index=X_test.index)

optimized_cost = (y_test * optimized_tariffs / 60).sum()

print(f"Actual cost:    {actual_cost:.2f}")
print(f"Optimized cost: {optimized_cost:.2f}")
print(f"Potential saving: {actual_cost - optimized_cost:.2f}")
```
